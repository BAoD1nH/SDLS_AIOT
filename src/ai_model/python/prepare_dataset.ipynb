```python
import os
import shutil
import random

# Thư mục chứa dataset gốc
SOURCE_DIR = 'C:/Users/Bao Dinh/Downloads/facerecognition/105_classes_pins_dataset'

# Tên folder cần gán nhãn "registered"
REGISTERED_FOLDER = 'pins_tom ellis'

# Thư mục output
OUTPUT_DIR = os.path.join(SOURCE_DIR, 'output_faces_sampled_500')
REGISTERED_DIR = os.path.join(OUTPUT_DIR, 'registered')
OTHER_DIR = os.path.join(OUTPUT_DIR, 'other')

# Tạo thư mục nếu chưa có
os.makedirs(REGISTERED_DIR, exist_ok=True)
os.makedirs(OTHER_DIR, exist_ok=True)

# Biến đếm
registered_count = 0
other_count = 0

# Danh sách lưu đường dẫn ảnh other
other_images = []

# Duyệt các thư mục con
for folder in os.listdir(SOURCE_DIR):
	folder_path = os.path.join(SOURCE_DIR, folder)
	if not os.path.isdir(folder_path):
		continue

	if folder.lower() == REGISTERED_FOLDER.lower():
		# Copy tất cả ảnh registered
		for file in os.listdir(folder_path):
			if file.lower().endswith(('.jpg', '.jpeg', '.png')):
				registered_count += 1
				src = os.path.join(folder_path, file)
				dst = os.path.join(REGISTERED_DIR, f'registered_{registered_count}.jpg')
				shutil.copy(src, dst)
	else:
		# Gom ảnh other
		for file in os.listdir(folder_path):
			if file.lower().endswith(('.jpg', '.jpeg', '.png')):
				full_path = os.path.join(folder_path, file)
				other_images.append(full_path)

# Shuffle và lấy 500 ảnh
random.shuffle(other_images)
sampled_others = other_images[:500]

# Copy ảnh sampled vào thư mục other
for i, src in enumerate(sampled_others, 1):
	dst = os.path.join(OTHER_DIR, f'other_{i}.jpg')
	shutil.copy(src, dst)
	other_count += 1

print(f"✅ Done: {registered_count} registered images, {other_count} other images sampled (max 500).")
```
